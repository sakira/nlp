### What is GloVe?

**Global Vectors for Word Representation**
- It is a global log-bilinear regression model [1].
- It is combining the advantage of  global matrix factorization [2] and local context window methods [3].

[1] Pennington, Jeffrey, Richard Socher, and Christopher D. Manning. "Glove: Global vectors for word representation." Proceedings of the 2014 conference on empirical methods in natural language processing (EMNLP). 2014.

[2] Deerwester, Scott, et al. "Indexing by latent semantic analysis." Journal of the American society for information science 41.6 (1990): 391-407.

[3] Mikolov, Tomáš, Wen-tau Yih, and Geoffrey Zweig. "Linguistic regularities in continuous space word representations." Proceedings of the 2013 conference of the north american chapter of the association for computational linguistics: Human language technologies. 2013.

### The GloVe model
- The main intuition is to combine word representation and language model.
- Uses weighted least square model.
- Trains global word-word co-occurrence counts.
